### Collaborative Filtering

Collaborative filtering is typically used for recommender systems. spark.mllib uses the alternating least squares (ALS) algorithm

In the following example rating data is loaded from an input file. Each row consists of a user, a product and a rating. The default ALS.train() method is used which assumes ratings are explicit. The recommendation is evaluated by measuring the Mean Squared Error of rating prediction.

```python
from pyspark.mllib.recommendation import ALS, MatrixFactorizationModel, Rating
```

Load the data

```python
data = sc.textFile("/home/mtatineni/dataset/als/test.data")
ratings = data.map(lambda l: l.split(','))\
    .map(lambda l: Rating(int(l[0]), int(l[1]), float(l[2])))
```

Build the model using ALS

```python
rank = 10
numIterations = 10
model = ALS.train(ratings, rank, numIterations)
```

Evaluate the model

```python
testdata = ratings.map(lambda p: (p[0], p[1]))
predictions = model.predictAll(testdata).map(lambda r: ((r[0], r[1]), r[2]))
ratesAndPreds = ratings.map(lambda r: ((r[0], r[1]), r[2])).join(predictions)
MSE = ratesAndPreds.map(lambda r: (r[1][0] - r[1][1])**2).mean()
print("Mean Squared Error = " + str(MSE))
```

K-means Example

```python
from numpy import array
from math import sqrt

from pyspark.mllib.clustering import KMeans, KMeansModel
```

Load data

```python
data = sc.textFile("/home/mtatineni/dataset/kmeans_data.txt")
parsedData = data.map(lambda line: array([float(x) for x in line.split(' ')]))
```

The k-means paramters are
- k is the number of desired clusters. 
- initializationMode specifies either random initialization or initialization via k-means.
- initializationSteps determines the number of steps in the k-means algorithm.
- epsilon determines the distance threshold within which we consider k-means to have converged.
- initialModel is an optional set of cluster centers used for initialization. If this parameter is supplied, only one run is performed.

Build model

```python
clusters = KMeans.train(parsedData, 2, maxIterations=10, initializationMode="random")
```

Evaluate

```python
def error(point):
    center = clusters.centers[clusters.predict(point)]
    return sqrt(sum([x**2 for x in (point - center)]))

WSSSE = parsedData.map(lambda point: error(point)).reduce(lambda x, y: x + y)
print("Within Set Sum of Squared Error = " + str(WSSSE))
```

LDA Example

Word count vectors are loaded. They represent a corpus of documents. LDA is used to infer three topics from the documents. Topics are printed out, represented as probability distributions over words.

```python
from pyspark.mllib.clustering import LDA, LDAModel
```

```python
from pyspark.mllib.linalg import Vectors
```

```python
data = sc.textFile("/home/mtatineni/dataset/sample_lda_data.txt")
parsedData = data.map(lambda line: Vectors.dense([float(x) for x in line.strip().split(' ')]))
```

Index

```python
corpus = parsedData.zipWithIndex().map(lambda x: [x[1], x[0]]).cache()
```

Cluster into 3 topics

```python
ldaModel = LDA.train(corpus, k=3)
```

Output topics

```python
print("Learned topics (as distributions over vocab of " + str(ldaModel.vocabSize())
      + " words):")
topics = ldaModel.topicsMatrix()
for topic in range(3):
    print("Topic " + str(topic) + ":")
    for word in range(0, ldaModel.vocabSize()):
        print(" " + str(topics[word][topic]))
```

### Dimensionality Reduction - RDD-based API

Singular value decomposition (SVD)

SVD factorizes a matrix into three matrices:1) an orthonormal matrix, whose columns are called left singular vectors, 2)  diagonal matrix with non-negative diagonals in descending order, whose diagonals are called singular values, and 3) orthonormal matrix, whose columns are called right singular vectors

```python
from pyspark.mllib.linalg import Vectors
from pyspark.mllib.linalg.distributed import RowMatrix
```

```python
rows = sc.parallelize([
    Vectors.sparse(5, {1: 1.0, 3: 7.0}),
    Vectors.dense(2.0, 0.0, 3.0, 4.0, 5.0),
    Vectors.dense(4.0, 0.0, 0.0, 6.0, 7.0)
])
```

```python
mat = RowMatrix(rows)
```

```python
svd = mat.computeSVD(5, computeU=True)
```

```python
s = svd.s 
```

```python
s
```

#### Principal component analysis (PCA)

Info from manpage

Principal component analysis (PCA) is a statistical method to find a rotation such that the first coordinate has the largest variance possible, and each succeeding coordinate in turn has the largest variance possible. The columns of the rotation matrix are called principal components. PCA is used widely in dimensionality reduction.

```python
from pyspark.mllib.linalg.distributed import RowMatrix

rows = sc.parallelize([
    Vectors.sparse(5, {1: 1.0, 3: 7.0}),
    Vectors.dense(2.0, 0.0, 3.0, 4.0, 5.0),
    Vectors.dense(4.0, 0.0, 0.0, 6.0, 7.0)
])
```

```python
mat = RowMatrix(rows)
```

```python
pc = mat.computePrincipalComponents(4)
```
