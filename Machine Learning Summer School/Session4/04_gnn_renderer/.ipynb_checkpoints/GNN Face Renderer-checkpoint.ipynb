## Example

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from scipy.misc import imresize, imsave

import networkx as nx
import numpy as np
import time
import os

from Generator import Generator

deconv_layer = 6 # 5 or 6
model_name = 'FaceGen.RaFD.model.d{}.adam'.format(deconv_layer)
#    model_path = '../de-id/generator/output/{}.h5'.format(model_name)
model_path = './models/{}.h5'.format(model_name) # locally stored models


renderer = Generator(model_path, deconv_layer=deconv_layer)

# possible emotions = {
# 0: 'neutral', 
#    1: 'anger', 
#        2: 'contempt',
#            3: 'disgust', 
#                4: 'fear', 
#                    5: 'happy', 
#                        6: 'sadness', 
#                            7: 'surprise'
# }

emotion = 'neutral' # 'happy' #'neutral'
identities = [i for i in range(1, 57)];
            
# generate and display artificial faces
everyone = [i for i in range(1, 57)]
single = [33]
k_anon = [53, 40, 33]

emotions = ['happy', 'neutral', 'sad']
identities = [single, k_anon, everyone]

fig = plt.figure(figsize=(15, 5))
fig.suptitle("Generated Faces", fontsize=16)

for ix, (ids, emo) in enumerate(zip(identities, emotions)):
    image = renderer.generate(ids, emo)
    ax = plt.subplot("13" + str(ix+1))
    ax.set_title("k = " + str(len(ids)) + ", Expression: " + emo)
    ax.imshow(image)
    
# show the generated faces
plt.show()
```
