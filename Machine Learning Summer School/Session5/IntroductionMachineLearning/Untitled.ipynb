# Curse of dimensionality #

The curse of dimensionality is one of the most important problems in the learning of multivariate machines. It appears in many different forms, but they all have the same source: the fact that points in high dimensional space are very sparse.

The simplest possible illustrative example is presented in one or two dimensions. Suppose we have two points on a line, 0 and 1. These two points are units of distance from each other. Let us suppose that we introduce a second axis of "data", again distributed to a distance from the unit. Now we have two points, (0.0) and (1.1).
But the distance between the points has grown to $\sqrt 2$.

In three dimensions the two points will be $\sqrt 3$ away, and so on. By the time we reached 10 dimensions, the two points are now at a distance of 3 from each other: three times as far away as they were when the data was one-dimensional.


```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np


plt.figure(figsize=(12, 6))
plt.title(
    "Euclidean Distance between $[0]^{1 x n}$ and $[1]^{1 x n}$ for $n \in \{1, \ldots, 10\}$"
)
plt.xlabel('$n$')
plt.ylabel('$|| a - b ||_2$')
plt.plot(range(1, 10), np.sqrt(range(1, 10)))
```
