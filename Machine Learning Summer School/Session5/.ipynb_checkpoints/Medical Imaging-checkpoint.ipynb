# Medical imaging

In the last 30 years, different advances have emerged in the area of medical imaging and computer processing. These have led to medical images of various modalities, which are considered a valuable resource for clinical diagnosis as they provide anatomical, physiological, metabolic and functional information of the human body. Among the most commonly used medical imaging modalities in medical diagnosis are x-ray imaging, computed tomography (CT), ultrasound imaging (US), single-photon emission computed tomography (SPECT), positron emission computed tomography (PET), magnetic resonance imaging (MRI), and microscopy images.

Depending on how medical images are generated, we can categorized then into the following groups:

|<img src="materials/modalities.png" alt="modalities" width="70%"\>|
|:--:| 
| *Medical imaging modalities. Source: https://www.imageclef.org/2012/medical* |

During the process of generating a medical image, the coordinates of the analyzed body part (object) are mapped to coordinates in the image. In the case of a linear imaging system, the correspondence between the real world (physical space where the object is located) and the image domain is made by simple linear transformations such as translation, rotation, and scaling. In this case, the mapping function is bijective, i.e. it can be used to restore the real world coordinates from the image coordinates. This is useful for co-registration between images, as well as for interpreting them. In the case of the mode of images produced by an external radiation source, the function $f(\alpha, \beta,\gamma)$, which represents the distribution of the radiation passing through the object, is mapped by the function $g(x,y,z)$ using an imaging system that transports the radiation to the image plane. In this way it is possible to obtain a projection of the 3D object in a 2D or 3D plane, according to the image plane and the imaging system used.

In the case of CT scans, data acquisition is a term used to describe the systematic collection of X-ray transmission readings from the patient. The acquisition components include the X-ray generator and X-ray tube, a beam-forming filter, pre-patient collimators, electronic detectors and ADCs.

A PACS is an image storage and distribution system. This definition corresponds to the literal translation of the acronym Picture Archiving and Communications System. Normally this system is associated with Radiology, because this service is the main image generator of a hospital and also the one with the highest consumption.
In a stricter sense, it could be considered as a radiological image storage system, normally received from the different modalities.
A PACS system consists of three subsystems connected by computer networks and controlled by specialized software: medical image capture, storage in archive structures and visualization of image studies on workstations. These subsystems are described as follows:

a) Medical imaging. Information from medical imaging equipment (X-ray, US, MRI, CT, etc.) is entered into the PACS in DICOM (Digital Imaging COmmunication on Medicine) format. The acquisition is generally made directly in those cases in which the images are already digital, except in those cases in which the image is not digitized and is made indirectly. In the latter case, a digitizing device is required. Once digitized, they must be sent to the PACS. For this, it is essential to use a communications network.
b) Storage. The PACS is the intelligent engine of the system. Once the images are stored, a server file system is created for management, where they are stored, as well as all access to the images.
c) Display of the images. For the correct visualization of the images contained in the PACS, it is necessary to have a workstation that allows the radiology specialist to perform the tasks of visualization, diagnosis and analysis of the medical images in order to later issue a report.

### DICOM

It is important to note that PACS systems use specific protocols to capture images such as the DICOM protocol. A DICOM file  contains associated information on the patient, the study, the series and information specific to each imaging mode. 

- Patient level: 
The patient level contains demographic information and patient identification data (e.g., name, history number, date of birth). Because there may be more than one study, the patient's level is the highest. 
- Level of study: 
This is the most important level of the information model. A study is the result of a request for a certain type of medical examination. In general, the level of study may contain sets of images of one or more modalities (thus coming from different equipment).
- Serial level: 
This level contains images of the same mode. In order for several images to be grouped together in a series, one of the two conditions must be met: (1) temporarily related images or (2) spatially related images (e.g. cuts of the same organ). This level includes information on the date and time of acquisition, as well as details of the type of examination and equipment.
- Image Level: 
This level is the lowest level of the information model. The acquisition and positioning information is contained in the image. Depending on the image acquisition system, the image level contains data for only one image or a collection of images acquired over a short period of time (multi-frame images).

### Visualization of medical images (CT scans)

The combination of two-dimensional projection radiography with three-dimensional scanning geometry, as well as advances in image processing algorithms, allows for the generation of a three-dimensional image known as computed tomography (CT).  Tomography, from the Greek "tomos" meaning cut or slice, consists of an image or more slices through the body, the whole of which provides a three-dimensional distribution of the attenuation of the X-rays per unit volume. Because the cuts are made according to the axial plane, CT is also often called a computed tomography scan. 

|<img src="materials/human_anatomy_planes.png" alt="human_anatomy_planes" width="40%"\> |
|:--:| 
| *Human Anatomy Planes*. Source: Wikipedia.    |

In CT scans, the voxel values represent the attenuation of the x-rays passing through the different materials. 
To make the display independent of the image parameters, the attenuation coefficients of a material ($\mu$) are normalized using the Hounsfield scale (HU). This scale is based on the attenuation of materials such as water ($\mu_{water}$) and air ($\mu_{air}$).

$$ H = 1000 \times \frac{\mu - \mu_{water}}{\mu_{water} - \mu_{air}} $$

|<img src="materials/hounsfield-scale_CT.png" alt="hounsfield-scale_CT" width="60%"\> |

```python
# Auxiliary Functions

def load_scan(path):
    """This functions loads all dicom images inside path"""
    slices = [dicom.read_file(path + '/' + s) for s in os.listdir(path)]
    slices.sort(key = lambda x: int(x.InstanceNumber))
    try:
        slice_thickness = np.abs(slices[0].ImagePositionPatient[2] - slices[1].ImagePositionPatient[2])
    except:
        slice_thickness = np.abs(slices[0].SliceLocation - slices[1].SliceLocation)
    for s in slices:
        s.SliceThickness = slice_thickness
    return slices

def get_pixels_hu(scans):
    """This function convert DN to Hounsfield Units"""
    image = np.stack([s.pixel_array for s in scans])
    image = image.astype(np.int16) # convert to 16 bits
    # Set outside-of-scan pixels to 1
    # Convert to Hounsfield units (HU)
    intercept = scans[0].RescaleIntercept
    slope = scans[0].RescaleSlope
    if slope != 1:
        image = slope * image.astype(np.float64)
        image = image.astype(np.int16)
    image += np.int16(intercept)
    return np.array(image, dtype=np.int16)
```

```python
# Plotting different slices

def sample_stack(stack, rows=6, cols=6, start_with=10, show_every=3):
    fig,ax = plt.subplots(rows,cols,figsize=[rows * 2, cols * 2])
    for idx in range(rows * cols):
        slice_id = start_with + idx * show_every
        idx_x = int(idx/rows)
        idx_y = int(idx % rows)
        ax[idx_x, idx_y].set_title('slice {}'.format(slice_id))
        ax[idx_x, idx_y].imshow(stack[slice_id], cmap='gray')
        ax[idx_x, idx_y].axis('off')
    plt.show()

sample_stack(imgs_to_process)
```

```python
imgs_to_process = np.load('lung_image.npy').astype(np.float64) 

def resample(image, scan, new_spacing=[1,1,1]):
    # Determine current pixel spacing
    spacing = map(float,([scan[0].SliceThickness, scan[0].PixelSpacing[0], scan[0].PixelSpacing[1]]))
    spacing = np.array(list(spacing))
    resize_factor = spacing / new_spacing
    new_real_shape = image.shape * resize_factor
    new_shape = np.round(new_real_shape)
    real_resize_factor = new_shape / image.shape
    new_spacing = spacing / real_resize_factor
    image = scipy.ndimage.interpolation.zoom(image, real_resize_factor)
    return image, new_spacing

print('Shape before resampling: {}'.format(imgs_to_process.shape))

# resampling ct-scans
imgs_after_resamp, spacing = resample(imgs_to_process, patient, [1,1,1])
print('Shape after resampling: {}'.format(imgs_after_resamp.shape))

if not os.path.exists('lung_image_resampled.npy'):
    np.save('lung_image_resampled.npy', imgs_after_resamp)
```

```python
import plotly

def make_mesh(image, threshold=-300, step_size=1):
    print("Transposing surface")
    p = image.transpose(2,1,0)
    print("Calculating surface")
    verts, faces, norm, val = measure.marching_cubes(p, threshold, step_size=step_size, allow_degenerate=True) 
    return verts, faces

def plotly_3d(verts, faces):
    x,y,z = zip(*verts) 
    print("Drawing")
    # Make the colormap single color since the axes are positional not intensity. 
    # colormap=['rgb(255,105,180)','rgb(255,255,51)','rgb(0,191,255)']
    colormap=['rgb(236, 236, 212)','rgb(236, 236, 212)']
    
    fig = plotly.figure_factory.create_trisurf(x=x,
                        y=y, 
                        z=z, 
                        plot_edges=False,
                        colormap=colormap,
                        simplices=faces,
                        backgroundcolor='rgb(64, 64, 64)',
                        title="Interactive Visualization")
    iplot(fig)

def plt_3d(verts, faces):
    print("Drawing")
    x,y,z = zip(*verts) 
    fig = plt.figure(figsize=(10, 10))
    ax = fig.add_subplot(111, projection='3d')

    # Fancy indexing: `verts[faces]` to generate a collection of triangles
    mesh = Poly3DCollection(verts[faces], linewidths=0.05, alpha=1)
    face_color = [1, 1, 0.9]
    mesh.set_facecolor(face_color)
    ax.add_collection3d(mesh)

    ax.set_xlim(0, max(x))
    ax.set_ylim(0, max(y))
    ax.set_zlim(0, max(z))
    ax.set_facecolor((0.7, 0.7, 0.7))
    plt.show()
```

```python
imgs_after_resamp = np.load('lung_image_resampled.npy')

# It takes a while
v, f = make_mesh(imgs_after_resamp, 350)
plt_3d(v, f)
```
