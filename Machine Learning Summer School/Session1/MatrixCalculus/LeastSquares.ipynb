# Least squares

The problem of least squares in this case will be defined to find, given the full-range matrix $A\in\mathbb{R}^{m\times n}$, and therefore, invertible and a vector $\vec{b}\in\mathbb{R}^{m\times1}$, the vector $\vec{x}\in\mathbb{R}^{n\times1}$ closest to the **column space** of matrix $A$, which we remember is denoted as $\mathcal{C}\left(A\right)$ and corresponds to the space generated by the columns of the matrix A, combined linearly by the components $x_ {i}$ of the vector $\vec{x}$:

$$\mathcal{C}\left(A\right)=\left\{ \vec{v}\in\mathbb{R}^{m}:\vec{v}=A\,\vec{x},\;\vec{x}\in\mathbb{R}^{n},\:A\in\mathbb{R}^{m\times n},\right\}$$

Assuming that $A$ is full-range and that $n<m$ is the projection of the vector $\vec{b}\in\mathbb{R}^{m\times1}$ **squared to simplify minimization** in the column space of matrix $A$ is given by:
$$
\begin{equation}
\textrm{proy}\left(\vec{b};A\right)=\textrm{argmin}_{\vec{v}\in\mathcal{C}\left(A\right)}\left\Vert \vec{v}-\vec{b}\right\Vert _{2}^{2}=\textrm{argmin}_{\vec{x}}\left(A\:\vec{x}-\vec{b}\right)\cdotp\left(A\:\vec{x}-\vec{b}\right)
\label{eq:least_squared}
\end{equation}
$$

It is then necessary to find the vector that minimizes equation $\left(A\:\vec{x}-\vec{b}\right)^{T}\left(A\:\vec{x}-\vec{b}\right)$, which is developed by:

$$\left(A\:\vec{x}-\vec{b}\right)^{T}\left(A\:\vec{x}-\vec{b}\right)=\left(\vec{x}^{T}\,A^{T}-\vec{b}^{T}\right)\left(A\:\vec{x}-\vec{b}\right)=\vec{x}^{T}\,A^{T}A\:\vec{x}-\vec{x}^{T}\,A^{T}\vec{b}-\vec{b}^{T}A\:\vec{x}+\vec{b}^{T}\vec{b}$$

Pay attention to terms $\vec{x}^{T}\,A^{T}\vec{b}$ and $\vec{b}^{T}A\:\vec{x}$. 
The first term corresponds to a product of matrices with dimensions $1\times n\,n\times m\,m\times1$ which results in a scalar, just like the second term associated with the product $1\times m\,m\times n\,n\times1$. This means that taking the transpose of the first term for example, the scalar remains the same, so then: $\left(\vec{x}^{T}\,A^{T}\vec{b}\right)^{T}=\vec{b}^{T}A\:\vec{x}$, since $\left(\vec{x}^{T}\,A^{T}\vec{b}\right)^{T}=\vec{b}^{T}\left(\vec{x}^{T}\,A^{T}\right)^{T}=\vec{b}^{T}A\,\vec{x}$.


Which simplifies the equation as follows:

$$\left(A\:\vec{x}-\vec{b}\right)^{T}\left(A\:\vec{x}-\vec{b}\right)=\vec{x}^{T}\,A^{T}A\:\vec{x}-2\,\vec{b}^{T}A\,\vec{x}+\vec{b}^{T}\vec{b}.$$

To perform the minimization:

$$f\left(\vec{x}\right)=\textrm{argmin}_{\vec{x}}\left(A\:\vec{x}-\vec{b}\right)\cdotp\left(A\:\vec{x}-\vec{b}\right)$$

The gradient of such dot product is computed and set equal to zero to find the minimum point. Remember that for every function that results in a scalar $f\left(\vec{x}\right)$, the gradient vector is calculated:

$$\nabla_{\vec{x}}\left(A\:\vec{x}-\vec{b}\right)^{T}\left(A\:\vec{x}-\vec{b}\right)=0$$
$$\Rightarrow\nabla_{\vec{x}}\left(\vec{x}^{T}\,A^{T}A\:\vec{x}-2\,\vec{b}^{T}A\,\vec{x}+\vec{b}^{T}\vec{b}\right)=0$$
$$\Rightarrow\nabla_{\vec{x}}\left(\vec{x}^{T}\,A^{T}A\:\vec{x}\right)-\nabla_{\vec{x}}\left(2\,\vec{b}^{T}A\,\vec{x}\right)+\nabla_{\vec{x}}\left(\vec{b}^{T}\vec{b}\right)=0$$

Note that for the gradient in the first term we have $\left(\vec{x}^{T}\,A^{T}A\:\vec{x}\right)=\left(\vec{x}^{T}\,K\:\vec{x}\right)$ with $K=A^{T}A$ a squared matrix and corresponds to the quadratic form, for which had already shown that the gradient is given by  $\nabla_{\vec{x}}\left(\vec{x}^{T}\,K\,\vec{x}\right)=2\,K\,\vec{x}$. Regarding the second term, it can be rewritten as $\left(2\,\vec{b}^{T}A\,\vec{x}\right)=\left(2\,\vec{k}^{T}\,\vec{x}\right)$, note that the product $\vec{b}^{T}A$ results in a vector with known values $\vec{b}^{T}A=\vec{k}^{T}\in\mathbb{R}^{1\times n}$ since the product terms are made with dimensions $1\times m\,m\times n$.

For a similar expression, leaving scalar 2 out, we already show that the gradient is given by $\nabla_{\vec{x}}\left(\vec{k}^{T}\vec{x}\right)=\vec{k}$. Due to the gradient rule $\nabla_{\vec{x}}\left(2\,\vec{b}^{T}A\,\vec{x}\right)=2A^{T}\vec{b}^{T}$. Finally, the third term corresponds to a constant, so that its gradient is zero, which leads to: $\Rightarrow2\,A^{T}A\,\vec{x}-2\,\vec{b}^{T}A=0$ and taking the transpose of the second scalar term: $$\Rightarrow2\,A^{T}A\,\vec{x}-2\,A^{T}\,\vec{b}=0$$ $$\Rightarrow A^{T}A\,\vec{x}=A^{T}\,\vec{b}$$ $$\Rightarrow\vec{x}=\left(A^{T}A\right)^{-1}A^{T}\,\vec{b}$$

## Code example:

```python
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np


fig = plt.figure()
fig2 = plt.figure()
fig3 = plt.figure()
ax = fig.gca(projection='3d')
ax2 = fig2.gca(projection='3d')
ax3 = fig3.gca(projection='3d')

# Make data.
X = np.arange(0, 2, 0.01)
Y = np.arange(0, 2, 0.01)
X, Y = np.meshgrid(X, Y)
#dot product for 2d vector
Z = np.multiply(X,X) + np.multiply(Y,Y);

# Plot the surface.
surf = ax.plot_surface(X, Y, Z)
# Customize the z axis.
ax.set_zlim(0, 5)

#analytical gradient
Xag = 2 * X;
Yag = 2 * Y;


#numerical gradient
h = 0.01 #dx step
gy, gx = np.gradient(Z, h)

# Plot the surface.
surf2 = ax2.plot_surface(X, Y, Xag)
# Customize the z axis.
ax2.set_zlim(0, 5)

# Plot the surface.
surf3 = ax3.plot_surface(X, Y, gx)
# Customize the z axis.
ax3.set_zlim(0, 5)
```
